```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(train_features, train_y)
```

```python
lin_reg.predict(train_features)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(lin_reg.predict(train_features), train_y)
```

```python
mean_squared_error(lin_reg.predict(test_features), test_y)
```
